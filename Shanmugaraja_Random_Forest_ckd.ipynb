```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report,f1_score,roc_auc_score
```

```python
data=pd.read_csv("CKD.csv")
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
data["classification_yes"].value_counts()
```

```python
independent=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=data[['classification_yes']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
param_grid={"n_estimators":[10,100],"criterion":["gini","entropy"],"max_features":["auto","sqrt","log2"]}
```

```python
grid=GridSearchCV(RandomForestClassifier(),param_grid,n_jobs=-1,scoring="f1",verbose=3)
grid.fit(x_train,y_train)
```

```python
dr=grid.cv_results_
table=pd.DataFrame(dr)
table
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
f1=f1_score(y_test,y_pred,average='weighted')
f1
```

```python
roc=roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
roc
```
