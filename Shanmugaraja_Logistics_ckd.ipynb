```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
data=pd.read_csv("CKD.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
data["classification_yes"].value_counts()
```

```python
data.columns
```

```python
independent=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
col=independent.loc[1]
```

```python
independent.shape
```

```python
dependent=data[['classification_yes']]
le=LabelEncoder()
dependent=le.fit_transform(dependent)
dependent
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
x_train
```

```python
sc1=StandardScaler()
x_train=sc1.fit_transform(x_train)
x_test=sc1.transform(x_test)
x_test
```

```python
param_grid = {"solver":['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
LogisticRegression()

grid = GridSearchCV(LogisticRegression(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 

grid.fit(x_train, y_train) 
```

```python
dr=grid.cv_results_
table=pd.DataFrame(dr)
table
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
f1=f1_score(y_test,y_pred,average='weighted')
f1
```

```python
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
